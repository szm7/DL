Question 1

```python


def costFunction(w):
    return w ** 3-  20 * w ** 2 + 100 * w

def gradient(w):
    return 3 * w **2 - 40 * w + 100

def momentum(lr, iter):
    w = 5
    b = 0

    for i in range(iter):
        grad = gradient(w)
        w = w - lr * grad

        revenue_loss = costFunction(w)
        print(f"Iteration {i+1}: Price = {w:.4f}, Revenue Loss = {revenue_loss:.4f}")

    return w

learningRate = 0.05
iteration = 3

optimal_price = momentum(learningRate, iteration)

print("\nOptimal Price:", optimal_price)
```

Question 2

```python



def sigmoidFunction(x):
    return 1 / (1 + np.exp(-x))

def reluFunction(x):
    return np.maximum(0, x)

def sigmoidReluFunc(x):
    return 1 / (1 + np.exp(-x)) * np.maximum(0, x)


inputs = np.array([-5, -1, 0, 1, 2])


outputs_sigmoid = sigmoidFunction(inputs)
outputs_relu = reluFunction(inputs)
outputs_custom = sigmoidReluFunc(inputs)

# Plotting the results
plt.figure(figsize=(10, 8))
plt.plot(inputs, outputs_sigmoid, label='Sigmoid')
plt.plot(inputs, outputs_relu, label='ReLU')
plt.plot(inputs, outputs_custom, label='Custom (Sigmoid * ReLU)')
plt.legend()
plt.show()
```
